### **3) load data**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = tf.keras.utils.image_dataset_from_directory('/content/drive/MyDrive/deep learning/data')
```

```python
data_iterator = data.as_numpy_iterator() 
###Returns an iterator which converts all elements of the dataset to numpy i.e array
```

```python
batch = data_iterator.next()
```

```python
fig, ax = plt.subplots(ncols=5, figsize=(20,20))
for idx, img in enumerate(batch[0][:5]):
    ax[idx].imshow(img.astype(int))
    ax[idx].title.set_text(batch[1][idx])
```

### **4) Scalling the data**

```python
data = data.map(lambda x,y : (x/255, y)) ### for scalling the data i.e to convert it into arrays ranging from 0 to 1
```

```python
data.as_numpy_iterator().next()
```

### **5)splitting the data**

```python
train_size = int(len(data)*.7)
val_size = int(len(data)*.2)
test_size = int(len(data)*.1)
```

```python
train_size
```

```python
test_size
```

```python
val_size
```

```python
train=data.take(train_size)
val=data.skip(train_size).take(val_size)
test=data.skip(train_size+val_size).take(test_size)
```

### **6)Building the model**

```python
train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
model=Sequential()
```

```python
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='BinaryCrossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

### **7) Train the data**

```python
model_history=model.fit(train, validation_data=val,epochs=20)
```

### **8) Plot the Performance**

```python
plt.figure(figsize=(5,5))
plt.plot(model_history.history['loss'], color='teal', label='loss')
plt.plot(model_history.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.plot(model.history.history['accuracy'], color='teal', label='accuracy')
plt.plot(model.history.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

### **9) Evaluate**

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
pre = Precision()
re = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator(): 
    X, y = batch
    yhat = model.predict(X)
    pre.update_state(y, yhat)
    re.update_state(y, yhat)
    acc.update_state(y, yhat)
```

```python
print(pre.result(), re.result(), acc.result())
```
